```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hitters = pd.read_csv('List_of_Major_League_Baseball_players_to_hit_for_the_cycle_2.csv')
```

```python
hitters.drop('Ref', axis = 1)
```

```python
hitters['League'].value_counts()
```

```python
team_counts = hitters['Team'].value_counts().sort_index(ascending=True)
counts = list(team_counts)
```

```python
teams = hitters['Team'].unique()
team_names = sorted(teams)
```

```python
from matplotlib.pyplot import figure

plt.figure().set_figwidth(12)
plt.bar(team_names, counts)
plt.xticks(rotation='vertical')
plt.title('Frequency of Cycles for a Team')
plt.xlabel('Teams')
plt.ylabel('Number of Cylces')

```

```python
red_sox_cycles = hitters.loc[hitters['Team'] == 'Boston Red Sox']
red_sox_cycles
```

```python
total_number_of_games_as_of_today = 236126
number_of_cycles = hitters.shape[0]
number_of_cycles/total_number_of_games_as_of_today*100
```

```python
number_of_red_sox_cylces = red_sox_cycles.shape[0]
number_of_red_sox_cylces/total_number_of_games_as_of_today*100
```

```python
cycles_in_july = hitters.loc[pd.DatetimeIndex(hitters['Date']).month == 7]
cycles_in_july.loc[pd.DatetimeIndex(cycles_in_july['Date']).day == 31]
```
